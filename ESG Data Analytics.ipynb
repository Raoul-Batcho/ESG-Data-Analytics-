## Importing necessary libraries and data

```python
#libraries to read and manipulate data
import numpy as np
import pandas as pd

#libraries to visualize data
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading csv file into a pandas Dataframe

df=pd.read_csv('ESGData.csv', index_col=[0,1,2,3])
df.head()
```

```python
df.info()
```

## Data Preparation

```python
df=df.dropna(how='all', axis='columns') # Drop the last column since it contains only null values
df.info()
```

```python
df.shape
```

```python
# Convert all columns name to integer
df.columns=df.columns.astype(int)
```

```python
# put all column name in the list year
year=list(df.columns)
```

```python
# Reset the index
df.reset_index(inplace=True)
```

```python
df.info()
```

```python
def country(country_name):

  ''' This function takes a country name as argument
      and returns a dataframe object with each indicator as a column'''

  df_country = df.loc[df['Country Name']==country_name]                                      # extract from df where country corresponds to the name given as argument
  df_country = df_country.melt(id_vars=['Indicator Code'], var_name='year', value_vars=year)
  df_country=df_country.groupby(['Indicator Code', 'year'])['value'].sum()                   # group df_country by 'Indicator code' and 'year'
  df_country=pd.DataFrame(df_country)
  df_country.reset_index(inplace=True)                                                       # Reset the index column
  indicator=df_country['Indicator Code'].unique()                                            # Extract unique values of each indicator
  df_country2=pd.DataFrame()
  df_country2['year']=pd.DataFrame(df_country.loc[df_country['Indicator Code']==indicator[0]]['year'].values)  # Extract from df_country the years according to indicator code
  df_country2['country']=country_name
  for ind in indicator:
     df_country2[ind]=df_country.loc[df_country['Indicator Code']==ind]['value'].values       # Create a dataframe with each indicator as a column
  return df_country2.round(1)                                                                 # Round the values
```

```python
# Extract data where country is Togo
Togo=country('Togo')
Togo.head()
```

```python
# Extract data where country is Sub-Saharan Africa
sub_sahara= country('Sub-Saharan Africa')
sub_sahara.head()
```

```python
# Extract data where country is North America

north_america = country('North America')
north_america.head()
```

```python
# Create a list of selected  indicators including year
cols=['year',\
      'country',\
      'EG.ELC.ACCS.ZS',\
      'NV.AGR.TOTL.ZS',\
      'NY.GDP.MKTP.KD.ZG',\
      'SE.XPD.TOTL.GB.ZS',\
      'GB.XPD.RSDV.GD.ZS',\
      'IP.JRN.ARTC.SC',\
      'SL.UEM.TOTL.ZS' ]

# Create a list of selected  indicators without year
num_cols=['EG.ELC.ACCS.ZS',\
      'NV.AGR.TOTL.ZS',\
      'NY.GDP.MKTP.KD.ZG',\
      'SE.XPD.TOTL.GB.ZS',\
      'GB.XPD.RSDV.GD.ZS',\
      'IP.JRN.ARTC.SC',\
      'SL.UEM.TOTL.ZS' ]
```

```python
# Create Togo_df with selected indicators columns including year
Togo_df=Togo[cols]
Togo_df.info()
```

```python
# Create sub_sahara_df with selected indicators columns including year
sub_sahara_df=sub_sahara[cols]
sub_sahara_df.info()
```

```python
# Create north_america with selected indicators columns including year
north_america=north_america[cols]
north_america.info()
```

```python
# Concatenate Togo_df,sub_sahara_df,north_america
concat=pd.concat([Togo_df,sub_sahara_df,north_america], axis=0)
concat
```
